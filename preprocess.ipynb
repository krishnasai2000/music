```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df = df.drop(['beats'], axis=1)
```

```python
df.head()
```

## Visualizing the features

```python
df.plot()
plt.title("Plotting Various Features")
```

# Preprocesing

```python
df['class_name'].unique()
```

```python
df['class_name'] = df['class_name'].astype('category')
df['class_label'] = df['class_name'].cat.codes
```

```python
lookup_genre_name = dict(zip(df.class_label.unique(), df.class_name.unique()))   
lookup_genre_name
```

```python
df['class_name'].unique()
```

```python
cols = list(df.columns)
cols.remove('label')
cols.remove('class_label')
cols.remove('class_name')
#df[cols]
```

# Data Splitting

```python
X = df.iloc[:,1:28]
y = df['class_label']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X1 = scaler.fit_transform(X)
```

```python
X1
```

```python
X
```

# Feature importance using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0, n_jobs=-1).fit(X1,y)
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
names = [X.columns.values[i] for i in indices]
plt.figure()
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), importances[indices])
plt.xticks(range(X.shape[1]), names, rotation=90)
plt.show()
```

```python
rf.score(X1, y)
```

# Feature importance using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0).fit(X1, y)
importances = dt.feature_importances_
indices = np.argsort(importances)[::-1]
names = [X.columns.values[i] for i in indices]
plt.figure()
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), importances[indices])
plt.xticks(range(X.shape[1]), names, rotation=90)
plt.show()
```

```python
dt.score(X1, y)
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X1, y)
knn.score(X1,y)
```
